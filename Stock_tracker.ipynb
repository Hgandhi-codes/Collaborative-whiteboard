```python
import gradio as gr
import yfinance as yf
import plotly.graph_objects as go
from datetime import datetime, timedelta

def get_stock_data(symbol):
    stock = yf.Ticker(symbol)
    end_date = datetime.now()
    start_date = end_date - timedelta(days=30)
    hist = stock.history(start=start_date, end=end_date)
    return hist

def plot_stock(symbol):
    try:
        data = get_stock_data(symbol)
        
        fig = go.Figure()
        fig.add_trace(go.Scatter(
            x=data.index, 
            y=data['Close'],
            mode='lines',
            name='Close Price',
            line=dict(color='blue', width=2)
        ))
        
        current_price = data['Close'].iloc[-1]
        
        fig.update_layout(
            title=f"{symbol.upper()} Stock Price - Last 30 Days",
            xaxis_title="Date",
            yaxis_title="Price ($)",
            hovermode='x unified'
        )
        
        return fig, f"Current Price: ${current_price:.2f}"
    except:
        return None, "Error: Invalid symbol or no data"

iface = gr.Interface(
    fn=plot_stock,
    inputs=gr.Textbox(label="Stock Symbol", placeholder="AAPL, TSLA, MSFT, GOOGL"),
    outputs=[gr.Plot(), gr.Textbox(label="Price Info")],
    title="Stock Price Tracker",
    description="Enter any stock symbol to see its 30-day price chart"
)

iface.launch()
```
